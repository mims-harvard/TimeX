```python
# Define masks manually:
mask1 = torch.zeros_like(X[:,-2,:].unsqueeze(1)).float()
mask1[60:85,0,0] = 1
mask1 = mask1.transpose(0, 1) # (B, T, d)

mask2 = torch.zeros_like(X[:,-2,:].unsqueeze(1))
mask2[103:125,0,0] = 1
mask2 = mask2.transpose(0, 1) # (B, T, d)
```

```python
# Now set PreChosenConceptList:
from txai.utils.concepts import PreChosenConceptList

times = D['train_loader'].times
Xclass = X[:,-2,:].unsqueeze(1).repeat(1, 2, 1)
times_class = times[:,-2].unsqueeze(1).repeat(1, 2)
mask_class_univariate = torch.cat([mask1.squeeze(-1), mask2.squeeze(-1)], dim = 0)
mask_class_multivariate = torch.cat([mask1, mask2], dim = 0)
print(mask_class_multivariate.shape)

concept_class_uni = PreChosenConceptList(X = Xclass.clone(), times = times_class.clone(), masks = mask_class_univariate)
#concept_class_multi = PreChosenConceptList(X = Xclass.clone(), times = times_class.clone(), masks = mask_class_multivariate)
```
