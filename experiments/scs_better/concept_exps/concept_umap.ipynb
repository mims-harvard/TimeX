```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
z_concepts = torch.load('embeddings/model1_embeds.pt')
```

```python
print(z_concepts.shape)
```

```python
import umap

full_c = z_concepts.flatten(0, 1)
full_c_np = full_c.numpy()
labels = np.array(([0] * 100) + ([1] * 100))
```

```python
zc_umap = umap.UMAP().fit_transform(full_c_np)
```

```python
print(zc_umap.shape)
```

```python
plt.figure(dpi=200)
colors = (['blue'] * 100) + (['orange'] * 100)
plt.scatter(zc_umap[:100,0], zc_umap[:100,1], c = 'blue', label = 'Increasing')
plt.scatter(zc_umap[100:,0], zc_umap[100:,1], c = 'orange', label = 'Decreasing')
plt.title('Concept Embeddings on Exp. Embedding Model')
plt.legend()
plt.show()
```

```python
model_out = torch.load('embeddings/exp1_test_outputs.pt')
print(model_out['all_z'][1].shape)
```

```python
exp_z = model_out['all_z'][1]
```

```python
allz = torch.cat([full_c, exp_z], dim = 0)
allz_np = allz.numpy()
```

```python
U = umap.UMAP()
U.fit(allz_np)
```

```python
zc_umap = U.transform(full_c)
ze_umap = U.transform(exp_z.numpy())
```

```python
plt.figure(dpi=200)
plt.scatter(zc_umap[:100,0], zc_umap[:100,1], c = 'blue', label = 'Increasing')
plt.scatter(zc_umap[100:,0], zc_umap[100:,1], c = 'orange', label = 'Decreasing')

plt.scatter(ze_umap[:250,0], ze_umap[:250,1], c = 'limegreen', label = 'Class 1')
plt.scatter(ze_umap[250:500,0], ze_umap[250:500,1], c = 'green', label = 'Class 2')
plt.scatter(ze_umap[500:,0], ze_umap[500:,1], c = 'darkgreen', label = 'Class 3')

plt.title('Concept Embeddings on Exp. Embedding Model')
plt.legend()
plt.show()
```
